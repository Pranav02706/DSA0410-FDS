```python
import pandas as pd
import numpy as np
```

```python
f=pd.read_csv("/content/House_Prediction - House_Prediction.csv")
location_price_average=f.groupby("zip_code")["price"].mean()
more_than_4beds=f[f["beds"]>4].count()["beds"]
largest_sq_feet=f["size"].max()
print("Average house price for each location: ", location_price_average)
print("Number of houses with more than 4 bedrooms: ", more_than_4beds)
print("Largest house size: ", largest_sq_feet)
```
