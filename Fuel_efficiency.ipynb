```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/Fuel_Efficiency datasets.xlsx - Sheet1.csv")
c1=df[df['make']=="mazda"]
c2=df[df['make']=="toyota"]
mazda_avg = c1['highway_mpg(Fuel Efficiency)'].mean()
toyota_avg = c2['highway_mpg(Fuel Efficiency)'].mean()

print("The average fuel efficiency of Mazda is:", mazda_avg)
print("The average fuel efficiency of Toyota is:", toyota_avg)
print("The Percentage increase between both the models is:",((toyota_avg - mazda_avg) / mazda_avg) * 100,"%")
```
