```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=pd.read_csv("/content/order.csv")
total_orders=f.groupby("Full Name")["Order Count"].sum()
avg_order_per_product=f.groupby("Items")["Order Total"].mean()
earliest_order=f["Order"].min()
latest_order=f["Order"].max()

print("Total number of orders made by each customer: ",total_orders)
print("Average order quantity for each product: ",avg_order_per_product)
print("Earliest order date:", earliest_order)
print("Latest order date:", latest_order)
```
