```python
import pandas as pd
import numpy as np
```

```python
f=pd.read_csv('/content/Sales - Sales.csv')
avg=f[f['Month_sales'].str.contains('2023')]['Price'].mean()
print("average Sales in 2023 =",avg)
```

```python
f['Month_sales_num'] = f['Month_sales'].str.split(' ').str[0].map({
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
})

Q1 = f[f['Month_sales_num'].isin([1, 2, 3])]['Total'].sum()
Q4 = f[f['Month_sales_num'].isin([10, 11, 12])]['Total'].sum()

total_sales_year = f['Total'].sum()

if Q1 != 0:
    percentage_increase = ((Q4 - Q1) / Q1) * 100
else:
    percentage_increase = 0

print("Total sales for the year:", round(total_sales_year, 2))
print("Percentage increase from Q1 to Q4:", round(percentage_increase, 2), "%")
```
